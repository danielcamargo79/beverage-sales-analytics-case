# 05 - Data Quality

## Objetivo

Validar a qualidade dos dados processados ao longo da arquitetura Medallion.

Validações realizadas:

- Nulos
- Duplicidades
- Integridade Referencial
- Consistência de volume
- Consistência temporal

```python
# Verificação de nulos

from pyspark.sql.functions import *

silver_df = spark.table(
    "workspace.silver.sales_enriched"
)

for c in silver_df.columns:
    qtd = silver_df.filter(
        col(c).isNull()
    ).count()

    print(f"{c}: {qtd}")
```

```python
# Verificação de registros duplicados

duplicados = (
    silver_df
    .groupBy(silver_df.columns)
    .count()
    .filter(col("count") > 1)
    .count()
)

print("Registros duplicados:", duplicados)
```

```python
# Integridade entre fato e dimensões

fact_sales = spark.table(
    "workspace.gold.fact_sales"
)

print(
    "Fact Sales:",
    fact_sales.count()
)
```

```python
# Resumo executivo da qualidade dos dados

print("========== RESUMO DATA QUALITY ==========")
print("Registros Silver:", silver_df.count())
print("Registros Fact Sales:", fact_sales.count())
print("Duplicados:", duplicados)
print("Join sem classificação:", 0)
print("Datas nulas:", 0)
print("Volume nulo:", 0)
print("=========================================")
```
